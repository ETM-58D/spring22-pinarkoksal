The results show that coin amount and unemployement rate has a correlation of 0.51. This is an unexpected result since we expect a decrease on the demand as the unemployement increase.

And there is a negative correlation of -0.1, between interest rate and coin amount.

```r
# #numerical statistics
summary_data=skim(demand)
print(summary_data)
```

Above is the statistical analysis of the data.

```r
#Plotting unemployement rate vs gold amount
plot(x = demand$Quarter.Coin.Amount, y= demand$Unemployement.Rate,
main="Unemployement Rate vs Gold Amount",ylab="Unemployement Rate",xlab="Gold Amount")
grid()
```

The graph above shows the unemployement rate and gold amount releation. There is variety and an increase in the gold amount as the unemployement rate increase as correlation coefficient indicated. Bu since it is not logical, it can be just a coincidence.

```r
#plotting interest rate vs gold amount
plot(x = demand$Quarter.Coin.Amount, y= demand$Monthly.Interest.Rate,
     main="Interest Rate vs Gold Amount",ylab="Interest Rate",xlab="Gold Amount")
```

And final graph shows interest rate vs gold amount. The data does not follow a specific pattern and it seems like there is not a strong correlation.

## Conclusion
Analysis show that there is a positive correlation between gold amount and unemployement rate. But I think both the unemployement and gold amount increase apart from each other due to the economic conditions in Turkey, and thus the results stated a positive correlation.

And expectedly, a negative correlation between intesest rate and gold amount is observed although it is not very strong.
